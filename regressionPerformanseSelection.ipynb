## Evaluating Regression Model Performance

### R-Squared

SS_res = Sum of Squares of residuals

1-SS_res/SS_tos   The nearer to 1, the best. The smallest the worst.

The problem using this parameter is that this value will never decrease if you add another variable.

That is why we want to use **Adjusted R^2**. It has a penalization factor to variables that do not help your model. So, only if when you add a variable the increase in R^2 is enough, and resists the penalization of the adjusted R^2, then you should add that variable to your model.

**Sklearn-Metrics** is the module that contains all the metrics used to evaluate all models in machine learning.
