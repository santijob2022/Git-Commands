## GIT Commands Basics

### Phase 1

A Git repository is a collection of objects and references. Objects have relationships with each other, and references point to objects and to other references. The main objects in a Git repository are commits, but other objects include blobs and trees. The most important references in Git are branches, which you can think of as labels you put on a commit.

HEAD is another important type of reference. The purpose of HEAD is to keep track of the current point in a Git repo. In other words, HEAD answers the question, “Where am I right now?”

**git init** -> Initialize repository on .git file.

**git add <file name>** -> first paste an archive adjacent to the file .git, then you can add it. 

**git status** -> to check that your archive created with **git add** is on the directory.

**git commit <name>** -> make changes to the respective archive then **commit** them to add the changes. Add the description.

**git pull** -> Gets the uploaded changes in the respective branches. 

**git push** -> Uploads your changes to the branch after commit.

**rm <*part_of_the_name.extension>** -> Removes existing archives. The symbol* is a wild card to eliminate archives faster.

**touch <name>** -> This creates a new archive, but you need to track them.

**git checkout** -> It is similar to **git status** however the changes should not have been committed to appear as a Modified file in the list given by **git checkou**.

**git checkout HEAD^** -> Checks the status of commits one step behind.

**git checkout HEAD~2** -> Checks 2 steps behind.

**git checkout @{-1}** -> return to the previously checked Head

**git log --oneline** -> See the pointer to the history of the messages added when a change was commited.

**git log --stats** -> Shows more info than git log --oneline

**git show reference** -> Check the details of the given commit reference.

**git checkout branch name** -> Return Head(pointer) to the master pointer. But be aware of not making changes to the files cause it won´t let you go there again, since the 'past' has changed.
